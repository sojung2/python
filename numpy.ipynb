```python
arr1 = np.array([[1,2,3],
                 [6,7,8]])
arr2 = np.array([[10,20,30],
                 [10,20,30]])
```

```python
print(arr1 + arr2)
print(arr2 / arr1)

print(arr1.sum(axis=0)) # 행방향합계 
print(arr1.sum(axis=1)) # 열방향합계
```

```python
print(arr1.mean()) #평균
print(arr1.std())  #표준편차
print(arr1.var())  #분산
print(arr1.max()) #최대값
print(arr1.min()) #최소값
```

이미지를 이용한 배열구조 이해하기 

```python
from PIL import Image # 이미지의 정보를 가져올 떄 사용 
import numpy as np  
import matplotlib.pyplot as plt  # 화면에 시각화를 해야하면 항상 기본
```

```python
#이미지 읽어오기
cat_tmp = Image.open('./cat1.jpg')
print(cat_tmp)
plt.imshow(cat_tmp)  #가로 x 축 세로 y 축 img는 0,0 부터 시작한다.
#img는 기본적으로 3차원의 배열구조를 가지고 있다. 
```

```python
imgArr =np.array(cat_tmp)
imgArr
```

```python
imgArr.shape #행 열 색상
```

고양이 이미지 잘라오기 

```python
#이미지 상단 자르기 
cat_top = imgArr[0:280, :,:]
plt.imshow(cat_top)
```

```python
#이미지의 중앙 오른쪽 출력 
cat_rig = imgArr[:,340:,:]
plt.imshow(cat_rig)
```

```python
cat_eye = imgArr[200:270,420:590,:]
plt.imshow(cat_eye)
```

```python
#고양이 눈 영역 색상변경 
#for문 사용. rgb값을 모두 50으로 변경 

#행위치 확인 
for i in range(len(imgArr)):
    if i > 200 and i <= 270:
        for j in range(len(imgArr[i])): #i번째 행 전체를 가져와 열 위치 확인
            if j >= 420 and j <= 590:
                for k in range(3): #RGB 색상변경
                    imgArr[i][j][k] = 50
                    
plt.imshow(imgArr)                    
```
